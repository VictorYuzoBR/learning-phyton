Crie uma lista de números e exiba o maior e o menor número.

```python
lista = []
loop = True
result = 0
while loop == True:
  x = input("Digite um número para por na lista ou sair para sair: ")
  if x == "sair":
    loop = False
  else:
    lista.append(x)

if len(lista) > 0:
  maior = lista[0]
else:
  print("lista sem itens")
  exit()

def maior(lista):

  result = lista[0]
  for i in lista:
    if float(i) > float(result):
      result = i
  return result

def menor(lista):

  result = float(lista[0])
  for i in lista:
    if float(i) < float(result):
      result = i
  return result

print(lista)
print("maior numero: " + str(maior(lista)))
print("menor numero: " + str(menor(lista)))
```

Implemente uma função para validar um CPF brasileiro.

```python
cpf = input("digite um cpf, apenas numeros")
cpfarray = []
primeirasoma = 0
segundasoma = 0
aux = 0
for i in cpf:
  if i.isdigit():
    cpfarray.append(int(i))

if len(cpfarray) != 11:
  print("cpf invalido")
  exit()

for i in range(10,1,-1):
  primeirasoma = primeirasoma + (cpfarray[aux] * i)
  aux += 1

divsoma = primeirasoma // 11
resto1 = primeirasoma - (divsoma * 11)
digito1 = 0
if resto1 <= 1:
  digito1 = 0
else:
  digito1 = 11 - resto1

if digito1 != cpfarray[9]:
  print("cpf invalido")
  exit()

aux = 0

for i in range(11,1,-1):
  segundasoma = segundasoma + (cpfarray[aux] * i)
  aux += 1

divsoma2 = segundasoma // 11
resto2 = segundasoma - (divsoma2 * 11)
digito2 = 0
if resto2 <= 1:
  digito2 = 0
else:
  digito2 = 11 - resto2

if digito2 != cpfarray[10]:
  print("cpf invalido")
  exit()

print("seu cpf é valido")

```

Remova duplicatas de uma lista de números.


```python
lista = [2,45,2,6,6,8,8,8,9,10]

print(lista)
tamanholista = len(lista)
aux = 0
for num in lista:
  for i in range(0,tamanholista):
    if i == aux:
      continue
    elif num == lista[i]:
       lista[i] = " "
  aux = aux+1

existerepetido = True
while existerepetido == True:
  tamanho = len(lista)
  aux = 0
  for i in range(0,tamanho):
    if lista[i] == " ":
      del lista[i]
      break
    else:
      aux = aux + 1
    if aux == tamanho:
      existerepetido = False
      break

print(lista)
```

Calcule a média e o desvio padrão de uma lista de números.

```python
import math
lista = [29,30,31,31.5,28,28.5,29]

soma = 0
for num in lista:
  soma = num + soma

divisao1 = soma / len(lista)

arraydesvio = []

for num in lista:
  desvio = (num - round(divisao1,2))**2
  arraydesvio.append(round(desvio,4))


somadesvio = 0
for num in arraydesvio:
  somadesvio = somadesvio + num

desviopadrao = math.sqrt(somadesvio / len(lista))

print(round(desviopadrao,4))
```

Conte quantas vezes cada número aparece em uma lista.

```python
lista = [1,2,3,3,4,4,4,5,5,8,8,8,8,8,8,8,8]
uniq = []
for num in lista:
  if num not in uniq:
    uniq.append(num)

for num in uniq:
  print(f"O numero {num} apareceu {lista.count(num)} vezes")
```

Implemente o algoritmo de ordenação por inserção (Insertion Sort).

```python
lista = [2,15,3,1]

for i in range(1,len(lista)):
  item = lista[i]
  aux = i-1
  while aux >= 0 and (item < lista[aux]):
    lista[aux + 1] = lista[aux]
    aux -= 1
    lista[aux+1] = item

print(lista)
```

Calcule a soma dos dígitos de um número inteiro fornecido.

```python
num = 33333
array = []
for i in str(num):
  array.append(i)

soma = 0

for x in array:
  numero = int(x)
  soma = soma + numero

print(soma)
```

Calcule a frequência de cada letra em uma string fornecida pelo usuário.

```python
frase = "esta eh uma string teste"

arrayletras = []
for p in frase:
  arrayletras.append(p)

uniq = []
for p in arrayletras:
  if p == " ":
    continue
  elif p not in uniq:
    uniq.append(p)
  else:
    continue

for letra in uniq:
  print(f"A letra {letra} aparece {arrayletras.count(letra)} vezes")

```
