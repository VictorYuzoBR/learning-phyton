```python
array = np.random.randint(1,10,size=100)
print(array)
print()

print(f"media: {np.median(array)}")

counts = {}
for i in array:
  counts.setdefault(i,0)
  counts[i] += 1


print()
moda = 0
maior = 0
for i,j in counts.items():
  if j > maior:
    maior = j
    moda = i


print(f"moda: {moda}")



```

```python
array1 = np.random.randint(1,10,size=3)
array2 = np.random.randint(1,10,size=3)

print(array1)
print(array2)
print()

aux = []
for i in range(0,3):
  if array1[i] > array2[i]:
    aux.append(array1[i]-array2[i])
  elif array2[i] > array1[i]:
    aux.append(array2[i]-array1[i])
  else:
    aux.append(0)

array3 = np.array(aux)
print(array3)
print(np.mean(array3))
```
