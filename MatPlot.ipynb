```python
x = np.arange(1,11)
y = np.sin(x)

fig , aux = plt.subplots()
aux.plot(x,y)
aux.axhline(0, color='black')
aux.fill_between(x, 1, where=y > 0, facecolor='green')
aux.fill_between(x, -1, where=y < 0, facecolor='red')
```

```python
x = np.arange(1,11)
y = 2 * x

plt.grid(True)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Grafico tabuada do 2")
plt.plot(x,y,linestyle="--", marker='o',color="red")
```

```python
x = np.arange(1,11)
y2 = 2*x
y1 = 3*x
y3 = 4*x
plt.grid(True)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Grafico tabuada do ")
plt.subplot(1,1,1)
plt.plot(x,y2,linestyle="--")
plt.subplot(1,1,1)
plt.plot(x,y1,linestyle="dotted")
plt.subplot(1,1,1)
plt.plot(x,y3)
```

```python
x = np.arange(1,11)
y2 = 2*x
y1 = 3*x

plt.grid(True)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Grafico tabuada do ")
plt.subplot(1,1,1)
plt.plot(x,y2,linestyle="--",label="linha 1")
plt.subplot(1,1,1)
plt.plot(x,y1,linestyle="dotted",label="linha 2")
plt.legend()
plt.show()
```

```python
x = np.arange(1,11)
y = x**2

plt.grid(True)
plt.grid(which='minor', color='#EEEEEE', linestyle=':', linewidth=0.5)
plt.minorticks_on()
plt.plot(x,y)
plt.show()
```
